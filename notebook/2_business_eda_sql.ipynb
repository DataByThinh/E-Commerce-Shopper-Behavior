```python
#Loading dataset 

import pandas as pd
import numpy as np
import duckdb

df = pd.read_csv("../data/raw/users_data.csv")
```

```python
engagement_table = duckdb.query("""

SELECT

    CASE
        WHEN product_views_per_day < 5 THEN 'Low Engagement'
        WHEN product_views_per_day < 15 THEN 'Medium Engagement'
        ELSE 'High Engagement'
    END AS engagement_segment,

    COUNT(*) AS users,

    ROUND(AVG(purchase_conversion_rate), 2) AS avg_conversion_rate,

    ROUND(AVG(monthly_spend), 2) AS avg_monthly_spend,

    ROUND(AVG(cart_items_average), 2) AS avg_cart_items

FROM read_csv_auto('../data/raw/users_data.csv')

GROUP BY 1

ORDER BY avg_conversion_rate DESC

""").to_df()

engagement_table
```
