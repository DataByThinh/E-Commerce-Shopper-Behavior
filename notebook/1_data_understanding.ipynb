```python
#Loading dataset 

import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/users_data.csv")
```

```python
# Checking null values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Checking unique values in each column
df.nunique().sort_values()
```
